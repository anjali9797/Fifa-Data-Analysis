# DATA VISUALIZATION

Plotting histograms on the basis of count of the players depending on the various attributes :

```python
df.describe()
```

From here we can see that the minimum age is 16 and maximum is 45. So we can plot histogram for these age groups using matplotlib

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
# plot histogram
ax.hist(df['Age'])
# set title and labels
ax.set_title('Histogram for Count of players for each age group ')
ax.set_xlabel('AGE')
ax.set_ylabel('Player Count')
```

Conclusion Drawn : From this histogram we can conclude that Maximum Players lie between the age group of 20 to 30.

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data = df['Age'].value_counts() 
# get x and y data 
points = data.index 
frequency = data.values 
# create bar chart 
ax.bar(points, frequency) 
# set title and labels 
ax.set_title('Bar Graph representing the count of players wrt their age') 
ax.set_xlabel('Age in Years') 
ax.set_ylabel('Count of Players')
```

The conclusion that we draw from this graph is the same as what we drew from the histogram.

Analysing the data on the basis of the clubs 

```python
dff1 = df.groupby("Club", as_index=False)["Potential","Overall"].mean()
dff1.sort_values(by="Potential",inplace=True,ascending=False)
ax=sns.catplot(y="Club",x="Potential",data=dff1.iloc[0:10,:],kind='bar',aspect=3,height=5)
plt.title("Top 10 Clubs with highest mean Potential",fontsize=20);
```

Having a clearer view :

```python
ax=sns.catplot(y="Club",x="Potential",data=dff1.iloc[0:10,:],aspect=3,height=5)
plt.title("Top 10 Clubs with highest mean Potential",fontsize=20);
```
